```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/WA_Fn-UseC_-Marketing-Customer-Value-Analysis.csv")

# 1. Eksik değer kontrolü
print("=== EKSİK DEĞERLER ===")
print(df.isnull().sum())

# 2. Hedef değişken dağılımı
print("\n=== TOTAL CLAIM AMOUNT İSTATİSTİKLERİ ===")
print(df['Total Claim Amount'].describe())

# 3. Hedef değişken histogramı
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
df['Total Claim Amount'].hist(bins=50)
plt.title('Total Claim Amount Dağılımı')

plt.subplot(1,2,2)
df['Total Claim Amount'].apply(lambda x: x+1).pipe(lambda s: __import__('numpy').log(s)).hist(bins=50)
plt.title('Log(Total Claim Amount) Dağılımı')
plt.tight_layout()
plt.show()

# 4. Kategorik kolonların unique değerleri
cat_cols = ['State', 'Coverage', 'Vehicle Class', 'Vehicle Size', 
            'Policy Type', 'EmploymentStatus', 'Location Code']
for col in cat_cols:
    print(f"\n{col}: {df[col].unique()}")
```
